```python
#Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
see class is a blueprint,template for creating objects while object is a real world entitiy.take car as a class with its particular type like bmw/mercedes etc based on diffeences in features. 
class car:
    def __init__(self,p,q):
        self.carname=p
        self.avg=q
    def get(self):
        return self.carname,self.avg
c=car('bmw',90)
print(c.get())

#Q2. Name the four pillars of OOPs.
they are abstraction,encapsulation,polymorphism and inheritance.


#Q3. Explain why the __init__() function is used. Give a suitable example.
it is used to initialised the class variables during object creation and is also a constructor.
class integer1:
    def __init__(self,p,q):
        self.a=p
        self.b=q
    def get(self):
        return self.a+self.b
c=integer1(23,90)
print(c.get())


#Q4. Why self is used in OOPs?
to bind the variables and functions  to the class.


#Q5. What is inheritance? Give an example for each type of inheritance.

to ensure reusability we inherit the class into another and access the the varibles using this class objects.
multilevel inheritance
class test1:
    def a(self):
        print('hi')
class test2(test1):
    def b(self):
        print('hello')
class test3(test2):
    pass
d=test3()
d.a()
d.b()

multiple inheritance
class test1:
    def a(self):
        print('bye')
class test2:
    def b(self):
        print('see u')
class test3(test1,test2):
    pass
d=test3()
d.a()
d.b()

single level
inheritance
class test:
    def abc(self):
        print('hello')
class child(test):
    pass
a=child()
a.abc()
```
